#Dataset: Students Exam Scores
Este dataset incluye los puntajes de tres diferentes pruebas de una escuela, relacionandolo con su contexto socioeconomico.

El **objetivo** de este proyecto es analizar las diferentes puntuaciones obtenidas por un grupo de estudiantes, y poder detectar a través del análisis cuales son los principales aspectos o factores que influyen en el rendimiento de tus puntajes obtenidos en sus examenes de diferentes asignaturas. Con la premisa de que los estudiantes o mejor dicho los jóvenes son el futuro parece relevante entender esto como sociedad, para poder prepararlos para los que les depara el futuro.
La **audiencia**, son principalmente el claustro docente, partiendo desde los maestros y los directores.
Este proyecto se basará en la recopilacion de datos de examenes de una muestra de estudiantes, con sus correspondientes datos socioeconómicos y culturales.
La **motivación** de este proyecto se basa en intentar comprender el porque de las diferencias entre los resultados de los alumnos, para verificar que acciones se pueden llevar a cabo para poder subsanar este problema.
